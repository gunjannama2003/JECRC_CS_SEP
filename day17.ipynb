```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x= np.linspace(0,5,11)
x1 = np.arange(1,11,3)
```

```python
x
```

```python
x1
```

```python
y= x**3
```

```python
plt.bar(x,y)
plt.title('Bar Graph')
plt.show()

plt.hist(x)
plt.title('Histogram')
plt.show()

plt.scatter(x,y)
plt.title('Scatter')

plt.show()

plt.violinplot(x)
plt.title('violinplot - badsoorat wala')
plt.show()

plt.boxplot(x,y)
plt.title('Boxplot')
plt.show()

plt.stem(x,y)
plt.title('Stem')
plt.show()


```

```python
y2  = x**4
```

```python
plt.plot(x, y , x , y2)
plt.legend(['y=x**3','y=x**4'])
plt.show()
```

```python
plt.plot(x, y )
plt.plot(x, y2)
plt.legend(['y=x**3','y=x**4'])
plt.show()
```

```python
fig = plt.figure(figsize = (10,5)) 
plt.plot(x, y ,lw=3 ,marker='o', markersize=15 ,markerfacecolor='yellow')
plt.plot(x, y2, marker = 'v', markersize=10)
plt.legend(['y=x**3','y=x**4'])
plt.show()
```
